# Importing Pandas, Numpy, Matplot and Seaborn 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
%pylab inline
```

# Loading the CSV file 

```python
uber_data=pd.read_csv('uber_raw_data.csv')
```

# Basic info about the data

```python
uber_data.head()
```

## Changing the Date/Time format to Pandas processing format  

```python
uber_data['Date/Time']=uber_data['Date/Time'].map(pd.to_datetime)
```

# This is how it look after changing the format

```python
uber_data.head()
```

# Adding the Days, Weekdays and the Hour column to the data

```python
def days_of_month(dt):
    return dt.day
uber_data['Days']=uber_data['Date/Time'].map(days_of_month)
```

```python
uber_data.tail()
```

```python
def weekdays(dt):
    return dt.weekday()
uber_data['Weekdays']=uber_data['Date/Time'].map(weekdays)
```

```python
uber_data.tail()
```

```python
def hours(dt):
    return dt.hour
uber_data['Hours']=uber_data['Date/Time'].map(hours)
```

```python
uber_data.tail()
```
